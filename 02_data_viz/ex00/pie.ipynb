```python
import psycopg2
import matplotlib.pyplot as plt
conn = psycopg2.connect(database="piscineds",
                        user='mpellegr', password='mysecretpassword', 
                        host='localhost', port='5432'
)
cursor = conn.cursor()
```

```python
cursor.execute('SELECT event_type, COUNT(*) as count FROM customers GROUP BY event_type')
res = cursor.fetchall()
data = []
labels = []
for event in res:
    data.append(event[1])
# print(res)
labels = 'view', 'cart', 'remove_from_cart', 'purchase'
fig, ax = plt.subplots()
ax.pie(sorted(data, reverse=True), labels=labels, autopct='%1.1f%%', wedgeprops={"linewidth": 1, "edgecolor": "white"})
plt.show()
```
